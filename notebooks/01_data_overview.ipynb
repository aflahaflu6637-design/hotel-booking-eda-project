### Purpose

The `describe()` function provides a statistical summary of numerical features. It helps understand the distribution, central tendency, spread, and variability of the data.

### Key Observations

- The average booking lead time is approximately **104 days**, indicating that customers generally book hotels well in advance.
- The average **ADR (Average Daily Rate)** is around **102**, representing the typical room price paid per booking.
- The average number of **adults per booking is close to 2**, suggesting that couples form a significant portion of hotel guests.
- Most bookings contain **no children or babies**, as indicated by the median values of zero.
- The majority of guests stay around **2–3 week nights** and **1 weekend night** on average.
- The cancellation rate is approximately **37%**, indicating that cancellations represent a significant business challenge.
- Several features show a large difference between minimum and maximum values, suggesting substantial variation in customer behavior and booking patterns.
- Some variables such as ADR, lead time, waiting days, and previous bookings contain extreme values that may require outlier analysis.

### Key Insight

The statistical summary reveals diverse booking behaviors, stay patterns, and pricing characteristics. Customers generally book several months in advance, travel mostly as couples, and exhibit varying booking habits. The presence of extreme values and high variability highlights the importance of further analysis and preprocessing before drawing final conclusions.
